## Line Plot

```python
from seaborn import relplot
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
```

```python
df.head()
```

```python
relplot(x = 'time', y = 'value', kind = 'line', data = df)
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
fig, ax = plt.subplots()
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, label = 'Sine')
```

```python
relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc='right')
```

```python
fmri[fmri['timepoint']==5].min()
```

```python
relplot(x = 'timepoint', y = 'signal', kind = 'line', estimator = None, data = fmri, 
        label = 'Signal', height = 4, aspect = 2)
plt.legend(loc='right')
```

```python
relplot(x = 'timepoint', y = 'signal', kind = 'line', hue = 'event', data = fmri, 
        label = 'Signal', height = 4, aspect = 2)
#plt.legend(loc='right')
```

```python
relplot(x = 'timepoint', y = 'signal', kind = 'line', hue = 'event', style = 'region',
        data = fmri, label = 'Signal', height = 4, aspect = 2)
#plt.legend(loc='right')
```

```python
relplot(x = 'timepoint', y = 'signal', kind = 'line', style = 'event', marker = True,
        dashes = False, data = fmri, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc='right')
```

```python
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```python
relplot(x = 'time', y = 'firing_rate', data = dots, kind = 'line')
```

```python
relplot(x = 'time', y = 'firing_rate', data = dots, kind = 'line', hue = 'coherence', style = 'choice')
```

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i*0.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sinplot(-1)
```

```python
sns.set_style('ticks', {'axes.grid': True, 'xticks.direction': 'in'})
sinplot()
```

```python
sns.set_style('ticks', {'axes.grid': True, 'xticks.direction': 'in'})
sinplot()
sns.despine()
```
